```python
import networkx as nx
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('adjacency.tsv', sep='\t')
```

```python
edge_list = []

for i, row in dataset.iterrows():
    edge_list.append((row[0], row[1]))
```

```python
graph = nx.from_edgelist(dataset.to_numpy())
```

```python
import matplotlib.pyplot as plt
```

```python
G = graph
plt.subplot(121)

nx.draw(G, pos=nx.circular_layout(G), node_size=0.01, linewidths=0.001, width=0.001)

plt.savefig("graph_circular.png", dpi=3600)
```

```python
G = graph
plt.subplot(121)

nx.draw(G, node_size=0.01, linewidths=0.001, width=0.001)

plt.savefig("graph_spring.png", dpi=3600)
```

```python
G = graph
plt.subplot(121)

nx.draw(G, pos=nx.spiral_layout(G), node_size=0.01, linewidths=0.001, width=0.001)

plt.savefig("spiral_circular.png", dpi=3600)
```

```python
G = graph
plt.subplot(121)

nx.draw(G, pos=nx.kamada_kawai_layout(G), node_size=0.01, linewidths=0.001, width=0.001)

plt.savefig("spiral_kamada_kawai.png", dpi=3600)
```

```python
degrees = [val for (node, val) in G.degree()]

plt.figure()
plt.hist(degrees, bins=100)
#plt.plot(list(range(0, max(degrees) + 1)), [i**(-3) for i in range(0, max(degrees) + 1)])
plt.title('Distribution of degrees of nodes')

degrees_counts = pd.Series(degrees).value_counts()

print('min:', min(degrees), 'max:', max(degrees), 'mean:', np.mean(degrees), 'std:', np.std(degrees))
```

```python
print('#nodes:', nx.classes.function.number_of_nodes(G), '#edges:', nx.classes.function.number_of_edges(G), '#connected components:', nx.algorithms.components.number_connected_components(G))
```

```python
avg_path_length = nx.average_shortest_path_length(G)

print('average shortest path length:', avg_path_length)
```

```python
number_nodes = nx.classes.function.number_of_nodes(G)
mean_degrees = np.mean(degrees)

print('BA aspl m=1:', np.log(number_nodes), 'BA aspl m>1:', np.log(number_nodes) / np.log(np.log(number_nodes)), 'WS aspl beta --> 0:', number_nodes/(2 * mean_degrees), 'WS aspl beta --> 1:', np.log(number_nodes) / np.log(mean_degrees))
```
