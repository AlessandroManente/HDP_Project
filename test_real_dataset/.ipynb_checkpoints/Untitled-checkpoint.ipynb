```python
import networkx as nx
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('adjacency.tsv', sep='\t')
```

```python
edge_list = []

for i, row in dataset.iterrows():
    edge_list.append((row[0], row[1]))
```

```python
graph = nx.from_edgelist(dataset.to_numpy())
```

```python
import matplotlib.pyplot as plt
```

```python
G = graph
plt.subplot(121)

nx.draw(G, pos=nx.circular_layout(G), node_size=2, linewidths=0.001, width=0.001)

plt.savefig("graph_circular.png", dpi=1200)
```

```python
G = graph
plt.subplot(121)

nx.draw(G, node_size=2, linewidths=0.001, width=0.001)

plt.savefig("graph_spring.png", dpi=1200)
```

```python
G = graph
plt.subplot(121)

nx.draw(G, pos=nx.spiral_layout(G), node_size=2, linewidths=0.001, width=0.001)

plt.savefig("spiral_circular.png", dpi=1200)
```
